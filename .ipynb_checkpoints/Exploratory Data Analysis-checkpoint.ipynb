```python
df[df['Continent'].str.contains('Oceania')]
```

```python
# Select columns from 1970 to 2022 (inclusive), reverse order
cols = ['1970 Population', '1980 Population', '1990 Population', 
        '2000 Population', '2010 Population', '2015 Population', 
        '2020 Population', '2022 Population']

df2 = df.groupby('Continent')[cols].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False) 
df2
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot()
```
